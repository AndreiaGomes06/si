### Exercise 12.2: Build a NN model for the breast-bin.csv dataset.

```python
breast_bin = r"C:\Users\Asus\Desktop\Bioinformática\2º Ano\1º Semestre\Sistemas inteligentes\si\datasets\breast-bin.csv"
bb_dataset = read_csv(breast_bin, sep=",", features = False, label=True)
```

```python
bb_dataset.X = StandardScaler().fit_transform(bb_dataset.X)
```

```python
bb_train, bb_test = train_test_split(bb_dataset)
```

```python
bb_l1 = Dense(input_size = 9, output_size = 9) 
bb_l2 = Dense(input_size = 9, output_size = 4)
bb_l3 = Dense(input_size = 4, output_size = 1)

bb_l1_activation = ReLUActivation()
bb_l2_activation = ReLUActivation()
bb_l3_activation = Sigmoid_Activation()

bb_model = NN(
    layers=[bb_l1, bb_l1_activation, bb_l2, bb_l2_activation, bb_l3, bb_l3_activation], verbose = True)
```

```python
bb_model.fit(dataset = bb_train)
```

```python
bb_model.predict(dataset = bb_test)
```

### Exercise 12.3: Build a NN model for the cpu.csv dataset.

```python
cpu = r"C:\Users\Asus\Desktop\Bioinformática\2º Ano\1º Semestre\Sistemas inteligentes\si\datasets\cpu.csv"
cpu_dataset = read_csv(cpu, sep=",", features=True, label=True)
```

```python
cpu_dataset.X = StandardScaler().fit_transform(cpu_dataset.X)
```

```python
cpu_model.fit(cpu_train)
```

```python
cpu_model.predict(dataset = cpu_test)
```
